```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from pydub import AudioSegment
import soundfile as sf
# Загрузка аудиофайлов
#audio1 = AudioSegment.from_file("Кот.aac", format="aac")
#audio2 = AudioSegment.from_file("машина.aac", format="aac")
#audio3 = AudioSegment.from_file("слон.aac", format="aac")

# Склеивание аудиофайлов
#combined_audio = audio1 + audio2 + audio3

# Загрузка аудиофайла
audio_file = "КОТ.aac"
y, sr = librosa.load(audio_file)

# Задать SNR в дБ
SNR_dB = 10
snr = 10.0**(SNR_dB/10.0)
# Мощность записанного аудиосигнала signal
p = y.var()
print(p)
# Требуемая мощность шума для заданного SNR
m = p/snr
# Шум с заданной мощностью (N – длина signal)
w = sp.sqrt(m)*sp.randn(len(y))
# Добавить шум к сигналу
s1 = y + w
# Применение алгоритма отделения речи от пауз
y_trim, index = librosa.effects.trim(s1,top_db=10)

# Нарисовать сигнал и границы слов
fig, ax = plt.subplots(nrows=3, sharex=True)
librosa.display.waveshow(y, sr=sr,ax=ax[0],color='r')
librosa.display.waveshow(s1, sr=sr,ax=ax[1],color='r')
librosa.display.waveshow(y_trim, sr=sr,ax=ax[2],color='r')
plt.show()
sf.write('noisy_audio_trimmed.wav', y_trim, sr)
```
